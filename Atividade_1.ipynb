```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

rng = np.random.default_rng(42)

# 1.
z = np.zeros(10)
print("1)", z, z.dtype, z.shape)

# 2.
u = np.ones(10, dtype=np.float64)
print("\n2)", u, u.dtype)

# 3.
seq = np.arange(10)
print("\n3)", seq)

# 4.
m_3 = np.arange(9).reshape(3, 3)
print("\n4)\n", m_3)

# 5.
v = np.array([1, 2, 0, 0, 4, 0])
idx_nz = np.nonzero(v)[0]
print("\n5)", idx_nz)

# 6.
lin = np.linspace(0, 1, 12)
print("\n6)", lin)

# 7.
I5 = np.eye(5)
print("7)\n", I5)

# 8.
r3 = rng.random((3, 3, 3))
print("\n8) shape", r3.shape)

# 9.
M = rng.random((10, 10))
print("\n9) min/max:", M.min(), M.max())

# 10.
x = np.array([1, 2, 3, 4, 5], dtype=float)
x_std = (x - x.mean()) / x.std(ddof=0)
print("\n10)", x_std, "mean=", x_std.mean(), "std=", x_std.std())

# 11.
arr = np.array([1, 2, 3], dtype=np.int32).astype(np.float32).astype(np.int64)
print("\n11)", arr, arr.dtype)

# 12.
a = np.array([2, 5, 7, 1, 9])
cl = np.clip(a, 0, 5)
print("\n12)", cl)

# 13.
rev = np.flip(np.arange(10))
print("\n13)", rev)

# 14.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
h = np.hstack([a, b])
vstack = np.vstack([a, b])
print("\n14) h=", h, "\n   v=\n", vstack)

# 15.
D4 = np.diag(np.full(4, 5.0))
print("\n15)\n", D4)

# 16.
A = np.arange(12).reshape(3, 4)
print("\n16) sum_rows=", A.sum(axis=1), "sum_cols=", A.sum(axis=0))

# 17.
integers = rng.integers(0, 10, size=100)
freq = np.bincount(integers, minlength=10)
print("\n17)", freq)

# 18.
v = np.arange(20)
sel = v[1::3]
print("\n18)", sel)

# 19.
vals = np.arange(20)
mask = (vals % 2 == 0)
print("\n19)", vals[mask])

# 20.
ll = np.array([[1, 2, 3], [4, 5, 6]])
print("\n20)", ll.shape, ll.dtype)

# 21.
dates = np.arange(np.datetime64('2025-01-01'), np.datetime64('2025-01-08'))
print("\n21)", dates)

# 22.
A = np.array([[1., 2., 3.], [4., 5., 6.]])
A_centered = A - A.mean(axis=0)
print("\n22)\n", A_centered)

# 23.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
elem = a * b
dotp = a @ b
print("\n23) elem=", elem, "dot=", dotp)

# 24.
A = np.arange(12).reshape(3, 4)
A[:, [0, 2]] = A[:, [2, 0]]
print("\n24)\n", A)
```

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

rng = np.random.default_rng(42)

# 25.
B = np.arange(6).reshape(2, 3)
r = B.ravel()
f = B.flatten()
r[0] = 99
print("25) B mod ravel:\n", B, "\nflatten é cópia? f[0]=", f[0])

# 26.
i, j = np.indices((8, 8))
board = (i + j) % 2
print("\n26)\n", board)

# 27.
x = np.arange(-5, 6)
y = np.where(x < 0, 0, x)
print("\n27)", y)

# 28.
big = rng.random(1_000_000)
mean_big = big.mean()
err = abs(mean_big - 0.5)
print("\n28) mean=", mean_big, "abs error=", err)

# 29.
a = np.array([1, 2, 3])[:, None]
b = np.array([10, 20, 30, 40])[None, :]
somas = a + b
print("\n29)\n", somas)

# 30.
def conv1d_valid(s, k):
    kf = np.flip(k)
    windows = sliding_window_view(s, k.size)
    return (windows * kf).sum(axis=1)

s = np.array([1, 2, 3, 4, 5])
k = np.array([1, 0, -1])
print("\n30)", conv1d_valid(s, k))

# 31.
A = rng.normal(size=(5, 4))
A_min = A.min(axis=0)
A_max = A.max(axis=0)
den = np.where(A_max > A_min, A_max - A_min, 1.0)
A01 = (A - A_min) / den
print("\n31)\n", A01)

# 32.
A = np.array([[3, 9], [1, 4], [2, 5]])
order = np.argsort(A[:, 1])
A_sorted = A[order]
print("\n32)\n", A_sorted)

# 33.
x = np.array([3, 1, 2, 3, 2, 3, 1])
vals, first_idx = np.unique(x, return_index=True)
counts = np.bincount(x, minlength=vals.max()+1)[vals]
print("\n33) uniques=", vals, "counts=", counts, "first_idx=", first_idx)

# 34.
P = rng.normal(size=(6, 2))
G = P @ P.T
sq = np.sum(P**2, axis=1, keepdims=True)
D2 = sq + sq.T - 2*G
D = np.sqrt(np.maximum(D2, 0.0))
print("\n34) dist shape:", D.shape)

# 35.
X = rng.standard_normal((1000, 3))
cov = np.cov(X, rowvar=False)
corr = np.corrcoef(X, rowvar=False)
print("\n35) cov shape", cov.shape, "corr shape", corr.shape)

# 36.
y = np.array([2, 0, 1, 2])
onehot = np.eye(y.max()+1)[y]
print("\n36)\n", onehot.astype(int))
```
